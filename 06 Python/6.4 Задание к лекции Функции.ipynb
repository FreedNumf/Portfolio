```python
# Исходные данные
documents = [
 {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
 {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
 {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]
directories = {
 '1': ['2207 876234', '11-2'],
 '2': ['10006'],
 '3': []
}
```

```python
# Пункт 1. Пользователь по команде "*p*" может узнать владельца документа по его номеру

def find_owner():
    """
    Пользователь по команде "*p*" может узнать владельца документа по его номеру
    """
    document = input('Введите номер документа:\n')
    for number_doc in documents:
        if document == number_doc["number"]:
            print(f'Владелец документа: {number_doc["name"]}')
            break
    else:
        print('Документ не найден в базе')
```

```python
# Пункт 2. Пользователь по команде "*s*" может по номеру документа узнать на какой полке он хранится

def shelf():
    """
    Пользователь по команде "*s*" может по номеру документа узнать на какой полке он хранится
    """
    doc_number = input('Введите номер документа:\n')
    for shelf_number in directories:
        if doc_number in directories[shelf_number]:
            print(f'Документ хранится на полке: ' + shelf_number)
            break
    else:
        print('Документ не найден в базе')
```

```python
def shelf2(num):
  for key, value in directories.items():
    for shel in value:
      if shel == num:
        return(key)
```

```python

# Пункт 3. Пользователь по команде "*l*" может увидеть полную информацию по всем документам
def full_inf():
    """
    Пользователь по команде "*l*" может увидеть полную информацию по всем документам
    """
    for key_doc in documents:
        num = key_doc.get('number')
        doc_type = key_doc.get('type')
        name = key_doc.get('name')
        num_shelf = shelf2(num)
        print(f"№: {num}, тип: {doc_type}, владелец: {doc_type}, полка хранения: {num_shelf}")
```

```python
def all_to_str():
    """Возвращаяет ключи словаряв в виде строки """
    return ", ".join(directories.keys())
```

```python
# Пункт 4. Пользователь по команде "*ads*" может добавить новую полку
def shelf_add():
    """
    Пользователь по команде "*ads*" может добавить новую полку
    """
    dir = input("Введите номер полки:\n")

    if dir in directories.keys():
        print(f'Такая полка уже существует.Текущий перечень полок: {all_to_str()}')
    else:
        directories[dir] = []
        print(f'Полка добавлена. Текущий перечень полок: {all_to_str()}')
```

```python
# Пункт 5. Пользователь по команде "*ds*" может удалить существующую полку из данных (только если она пустая)
def shelf_del():
    """
    Пользователь по команде "*ds*" может удалить существующую полку из данных (только если она пустая)
    """
    shelf_num = input('Введите номер полки:')
    res = directories.get(shelf_num, str())
    if shelf_num in directories.keys():
        if res:
            print(f'На полке есть документы, удалите их перед удалением полки. Текущий перечень полок: {all_to_str()}')
        else:
            directories.pop(shelf_num)
            print(f'Полка удалена. Текущий перечень полок: {all_to_str()}')
    else:
        print(f'Полки не существует! Текущий перечень полок: {all_to_str()}')
```

```python
# Меню программы
def main_menu():
    cmd = ''
    while cmd != 'q':
        cmd = input('Введите команду:\n')
        if cmd == "p":
            find_owner()
        elif cmd == 's':
            shelf()
        elif cmd == 'l':
            full_inf()
        elif cmd == 'ads':
            shelf_add()
        elif cmd == 'ds':
            shelf_del()
        elif cmd == 'q':
            print('Завершение работы.')
            break
        else:
            print('Команда не найдена')
```

```python
# Запуск программы
main_menu()
```
