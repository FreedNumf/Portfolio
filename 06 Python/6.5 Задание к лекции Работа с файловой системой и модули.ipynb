1. Переведите содержимое файла purchase_log.txt в словарь purchases вида:
```python
{'1840e0b9d4': 'Продукты', ...}
```

2. Для каждого user_id в файле visit_log.csv добавьте третий столбец с категорией покупки (если покупка была).

3. Запишите в файл funnel.csv визиты из файла visit_log.csv, в которых были покупки с указанием категории.

Учтите условия на данные:
- содержимое purchase_log.txt помещается в оперативную память компьютера
- содержимое visit_log.csv - нет; используйте только построчную обработку этого файла

````python
import json

purchases={}
with open('purchase_log.txt', encoding='utf-8') as p :
    for line in p:
        line = line.strip()
        d = json.loads(line)
        key = d['user_id']
        value = d['category']
        if key != 'user_id':
            purchases.setdefault(key, value)

with open('visit_log.csv', 'r') as v, open('funnel.csv', 'w') as f:
    for row in v:
        line_list = row.strip().split(',')
        if line_list[0] in purchases.keys():
            line_list.append(purchases[line_list[0]])
            add_line = ','.join(line_list)
        elif line_list[0] == 'user_id':
            line_list.append('category')
            add_line = ','.join(line_list)
        else:
            add_line = ','.join(line_list)
        f.write(add_line+'\n')
````
